```python
import numpy as np
import matplotlib.pyplot as plt

def filter_nan(x, y):
    """Remove NaN values from x and y arrays to avoid line breaks."""
    mask = ~np.isnan(x) & ~np.isnan(y)
    return x[mask], y[mask]

# Increase font sizes
#plt.rcParams.update({'font.size': 14, 'axes.labelsize': 16, 'axes.titlesize': 18, 'xtick.labelsize': 14, 'ytick.labelsize': 14, 'legend.fontsize': 14})

# Define save path
PLOT_SAVE_PATH = "/Users/sebastiengiroud/Documents/ETH/PhD/vscode/article2/figures/geochemistry_full_plot.pdf"

# Depths
depths_gas = np.array([-28, -201, -276, -516])

# Tritium (TU) with errors
tritium = np.array([5.7, 1.4, 1.1, 0.9])
tritium_err = np.array([0.1, 0.6, 0.1, 0.1])

# Noble gases (mol/L)
He4 = np.array([9.3e-3, 4.6, 7.1, 8.0])
Ne20 = np.array([8.4e-3, 7.5e-3, 1.0e-2, 7.8e-3])
Ar40 = np.array([16.3, 17.8, 21.1, 16.5])
Kr86 = np.array([6.4e-4, 6.6e-4, 7.3e-4, 6.0e-4])
Xe136 = np.array([4.7e-5, 4.9e-5, 5.0e-5, 4.2e-5])

# Isotopic ratios
He3_He4 = np.array([3.3e-7, 1.5e-8, 2.33e-08, 2.8e-8])
He3_He4_err = np.array([0.1e-7, 0.6e-8, 2e-08, 0.2e-8])
He3_He4_atm = 1.399e-6
Ar40_Ar36 = np.array([296.9, 320.2, 321.2, 320.6])
Ar40_Ar36_err = np.array([0.3, 0.3, 0.3, 0.3])
Ar40_Ar36_atm = 295.5

# Additional points
wells = ['AP', 'P11', 'S7', 'S9', 'S8', 'P14', 'P12', 'P201', 'P13', 'P205', 'P280', 'P600']
depths = np.asarray([-28.0, -30.0, -53.0, -70.0, -80.0, -108.0, -170.0, -201.0, -203.0, -250.0, -276.0, -516.0], dtype=np.float64)
temperature = np.asarray([22.2, 17.7, 18.2, 31.2, 25.4, 22.4, 29.5, 52.85, 35.2, 41.3, 53.2, 63.375], dtype=np.float64)
conductivity = np.asarray([340.0, 414.0, 478.0, 865.0, 586.0, 685.0, 495.0, 1488.25, 396.0, 1150.0, 1749.0, 1916.333333], dtype=np.float64)
pH = np.asarray([7.95, np.nan, 7.8, 7.9, 8.0, 8.0, np.nan, 7.37, np.nan, 8.0, 7.2, 7.27], dtype=np.float64)
filtered_pH, filtered_depths = filter_nan(pH, depths)



# Geochemical parameters
Ca = np.asarray([1.22, 1.42, 1.77, 1.34, 1.47, 0.94, 1.52, 1.46, 0.80, 1.08, 1.84, 1.37], dtype=np.float64)
Mg = np.asarray([0.46, 0.25, 0.37, 0.28, 0.51, 0.09, 0.34, 0.14, 0.25, 0.17, 0.17, 0.054], dtype=np.float64)
Na = np.asarray([0.59, 1.07, 0.84, 5.22, 2.27, 4.78, 1.59, 11.44, 1.74, 9.44, 12.79, 15.77], dtype=np.float64)
K = np.asarray([0.053, 0.062, 0.064, 0.144, 0.095, 0.105, 0.085, 0.261, 0.077, 0.172, 0.3, 0.342], dtype=np.float64)
Li = np.asarray([0.020, np.nan, 0.029, 0.174, 0.072, 0.101, np.nan, 0.372, 0.072, 0.261, 0.432, 0.537], dtype=np.float64)
SO4 = np.asarray([0.62, 0.72, 0.96, 2.03, 1.31, 2.16, 1.03, 4.35, 0.74, 3.35, 5.22, 5.82], dtype=np.float64)

geochemistry = [Ca, Mg, Na, K, Li, SO4]
geochemistry_labels = ["Ca", "Mg", "Na", "K", "Li", "SO4"]
symbols = ['o-', 's-', '^-', 'd-', 'x-', '*-']
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# Create subplots
fig, axs = plt.subplots(1, 5, figsize=(18, 6), sharey=True)

# Stable isotopes plot
""" ax = axs[0]
ax.plot(d18O, depths_gas, 'o-', label="-18O ()", color="blue")
ax.plot(d2H, depths_gas, 's-', label="-2H ()", color="green")
ax.set_xlabel("Stable Isotopes ()")
ax.set_ylabel("Depth (m)")
ax.legend()
ax.grid(True) """

# Tritium plot
ax = axs[0]
ax.errorbar(tritium, depths_gas, xerr=tritium_err, fmt='o-', color="red", label="Tritium (TU)")
ax.set_xlabel("Tritium (TU)")
ax.set_ylabel("Depth [m]")
ax.legend()
ax.grid(True)

# Noble gases plot
ax = axs[1]
ax.plot(He4, depths_gas, 'o-', label="He", color="purple")
ax.plot(Ne20, depths_gas, 's-', label="Ne", color="orange")
ax.plot(Ar40, depths_gas, '^-', label="Ar", color="brown")
ax.plot(Kr86, depths_gas, 'x-', label="Kr", color="cyan")
ax.plot(Xe136, depths_gas, 'd-', label="Xe", color="black")
ax.set_xlabel("Noble Gases (mol/L)")
ax.set_xscale("log")
ax.legend()
ax.grid(True)


# Isotopic ratios plot
ax = axs[2]
ax.plot(He3_He4/He3_He4_atm, depths_gas, 'o-', label="He/He", color="magenta")
ax.plot(Ar40_Ar36/Ar40_Ar36_atm, depths_gas, 's-', label="Ar/Ar", color="blue")
ax.set_xlabel("Isotopic Ratios (norm. to atm. Ra)")
ax.legend()
ax.grid(True)

# Temperature, pH & Conductivity Plot
ax = axs[3]
ax.plot(temperature/temperature[0], depths, 'o-', label="Temperature", color="red")
#ax.plot(pH/pH[0], depths, 's-', label="pH", color="green")
ax.plot(filtered_pH / filtered_pH[0], filtered_depths, 's-', label="pH", color="green")
ax.plot(conductivity/conductivity[0], depths, 'd-', label="Conductivity", color="blue")
ax.set_xlabel("T / pH / Conductivity (norm. to AP)")
ax.legend()
ax.grid(True)

# Geochemistry Plot
ax = axs[4]
for param, label, symbol, color in zip(geochemistry, geochemistry_labels, symbols, colors):
    filtered_param, filtered_depths = filter_nan(param, depths)
    ax.plot(filtered_param / filtered_param[0], filtered_depths, symbol, label=label, color=color)
ax.set_xlabel("Geochemical Parameters (norm. to AP)")
ax.legend()
ax.grid(True)

# Show and save plot
plt.tight_layout()
plt.savefig(PLOT_SAVE_PATH, bbox_inches='tight')
plt.show()

print(f"Figure saved to {PLOT_SAVE_PATH}")
```
